```python
import matplotlib.pyplot as plt

def max_rent(n, rents):
    if n == 0:
        return 0

    dp = [0] * (n+1)
    dp[1] = rents[0]

    for i in range(2, n+1):
        dp[i] = max(dp[i-1], dp[i-2] + rents[i-1])

    return dp[n]

# Membaca masukan
n = int(input())
rents = list(map(int, input().split()))

# Memanggil fungsi
result = max_rent(n, rents)
print(result)


# input sizes dan array untuk menyimpan running time
input_sizes = list(range(0, 1001, 10))
running_times = []

for n in input_sizes:
    import time
    start_time = time.time()
    max_rent(n, [i for i in range(1, n+1)])
    end_time = time.time()
    running_times.append(end_time - start_time)

plt.plot(input_sizes, running_times)
plt.xlabel('Input Size (n)')
plt.ylabel('Running Time (seconds)')
plt.title('Running Time vs. Input Size')
plt.show()
```
